```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
frames=[train_data.drop(['ID','y'],axis=1),test_data.drop('ID',axis=1)]
dataset=pd.concat(frames)
```

```python
categorical_features=[features for features in dataset.columns if dataset[features].dtypes=='O']
numerical_features=[features for features in dataset.columns if dataset[features].dtypes!='O']
```

```python
dataset['X75'].value_counts().count()
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder();

for i in categorical_features:
    dataset[i]=encode.fit_transform(dataset[i])
    
```

```python
from sklearn.preprocessing import StandardScaler
standarized_data = StandardScaler().fit_transform(dataset)
```

```python
# Import tSNE Analysis Algorithm
from sklearn.manifold import TSNE
model = TSNE(n_components=3, random_state=0,perplexity=50, n_iter=5000)
data = model.fit_transform(standarized_data)
```

```python
data.shape
```

```python
data=pd.DataFrame(data,columns =['1','2','3'])
data.head()
```

```python
X_train=data.iloc[:4209,:]
X_test=data.iloc[4209:,:]
```
